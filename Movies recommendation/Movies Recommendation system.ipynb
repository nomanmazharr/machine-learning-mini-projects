```python
import ast
def convert(text):
    ''' In this function we are taking only name key values from different columns '''
    data = []
    for i in ast.literal_eval(text):
        data.append(i['name'])# we can't apply it as it is string indices need to be integers thats y we take text in above form
    return data
```

```python
movies1['genres'] = movies1['genres'].apply(convert)
```

```python
movies1['keywords'] = movies1['keywords'].apply(convert)
```

```python
movies1['cast'] = movies1['cast'].apply(convert)
```

```python
movies1['cast'] = movies1['cast'].apply(lambda x:x[0:3])
```

```python
movies1.head()
```

```python
def fetch_drctr(text):
    ''' This helps us to get director name'''
    data = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            data.append(i['name'])
    return data
```

```python
movies1['crew'] = movies1['crew'].apply(fetch_drctr)
```

```python
movies1.head()
```

```python
def remove_space(text):
    '''This helps us to remove spaces from our lists of different columns in movies1'''
    data = []
    for i in text:
        data.append(i.replace(" ",''))
    return data
```

```python
movies1['crew'] = movies1['crew'].apply(remove_space)
movies1['genres'] = movies1['genres'].apply(remove_space)
movies1['keywords'] = movies1['keywords'].apply(remove_space)
movies1['cast'] = movies1['cast'].apply(remove_space)
```

```python
movies1.head()
```

```python
movies1['overview'] = movies1['overview'].apply(lambda x:x.split())
movies1['overview']
```

```python
movies1.head()
```

```python
movies1['tags'] = movies1['crew'] + movies1['cast'] + movies1['genres'] + movies1['keywords'] + movies1['overview']
```

```python
movies2 = movies1.drop(['crew','cast','genres','keywords','overview'],axis=1)
movies2.head()
```

```python
# It is used to remove all commas or other signs and gives spaces in words
movies2['tags'] = movies2['tags'].apply(lambda x:" ".join(x)) 
```

```python
movies2.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vector = CountVectorizer(max_features=5000,stop_words='english')
new_data = vector.fit_transform(movies2['tags']).toarray()
```

```python
new_data
```

```python
from sklearn.metrics.pairwise import cosine_similarity
movies_data = cosine_similarity(new_data)
movies_data
```

```python
movies2[movies2['title'] ==  'Newlyweds'].index[0]  # This value gives us the index of movie by using its title
```

```python
def recommend(movie):
    index = movies2[movies2['title'] == movie].index[0]
    distances = sorted(list(enumerate(movies_data[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]: # This for loop gives us first 6 values from our movies_data which is similar movies
        print(movies2.iloc[i[0]].title)  # here we get titles of similar movies of our given movie 
```

```python
recommend('Avatar')
```
