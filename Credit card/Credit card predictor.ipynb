```python
import pandas as pd
```

```python
credit_card = pd.read_csv('creditcard.csv')
```

```python
legit_credit_card = credit_card[credit_card.Class==0]
fraud_credit_card = credit_card[credit_card.Class==1]
```

```python
legit_credit_card.head()
```

```python
fraud_credit_card.head()
```

```python
legit_credit_card = legit_credit_card.sample(n=492)
```

```python
fraud_credit_card.shape
```

```python
legit_credit_card.shape
```

```python
data = pd.concat([legit_credit_card,fraud_credit_card],axis=0)
```

```python
data.shape
```

```python
data.head()
```

```python
data.reset_index()
```

```python
data.shape
```

```python
x = data.drop('Class',axis=1)
y = data['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
predict = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(predict,y_test)
```
